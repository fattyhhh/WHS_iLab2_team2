```python
import pandas as pd

# Merge the data frames on the 'name' column
merged_df = pd.merge(df1, df2, on='name', how='left')

df1['abn_look_up'] = merged_df['abn_look_up']
df1['location'] = merged_df['location']

# Print the updated d1 data frame
df1
```

```python
df1['web_content'] = df1['text'] + df1['sublinkstext']
```

```python
df1.drop(columns=['text','sublinkstext','sublinks'],inplace=True)
```

```python
df1
```

```python
df_cleaned = df1.dropna()
```

```python
df_cleaned
```

```python
def Abn_search(keywords):
    results = []
    for keyword in keywords:
        # Search 'web contents' column for keywords and extract corresponding data
        filtered_data = df1[df1['web_content'].str.contains(keyword, case=False, na=False)]
        for index, row in filtered_data.iterrows():
            results.append({
                'abn': row['abn_look_up'],
                'name' : row['name'],
                'location': row['location'],
                'website': row['website']
            })
    # Convert the list of dictionaries to a DataFrame
    results_df = pd.DataFrame(results)
    return results_df
```

```python
# Example usage of the function
keywords = ['renovation']
search_results = Abn_search(keywords)

# Print the search results
search_results
```
